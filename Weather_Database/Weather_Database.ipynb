```python
# Import necessary dependencies for task
import json 
import requests
from citipy import citipy
import pandas as pd
import numpy as np

from datetime import datetime
from config import weather_api_key
```

```python
# Create a new set of 2,000 random latitudes and longitudes.

# Create a new list called coordinates
# Random variations of 2,000 lats and lngs
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)

# Using ZIP fucntion to extrract list tuple
lat_lngs = zip(lats, lngs)
lat_lngs
```

```python
# Variations added to new list
coordinates = list(lat_lngs)
print (coordinates[:10])
```

```python
# Get the nearest city using the citipy module.

cities = []

for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name

    # filter only unique cities and add to "cities" list
    if city not in cities:
        cities.append(city)
```

```python
# Checking city count
len(cities)
```

```python
# Perform an API call with the OpenWeatherMap.
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
```

```python
# Retrieve the following information from the API call:

# Latitude and longitude
# Maximum temperature
# Percent humidity
# Percent cloudiness
# Wind speed
# Weather description (for example, clouds, fog, light rain, clear sky)


city_data = []

print("Beginning Data Retrieval     ")
print("-----------------------------")

# Counters.
record_count = 1
set_count = 1

# For loop for cities list
for i, city in enumerate(cities):

    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
    # URL
    city_url = url + "&q=" + city.replace(" ","+")

    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    
    record_count += 1
    
    # API requests for cities.
    try:
        # Data retreival and JSON Parse
        city_weather = requests.get(city_url).json()
       
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_country = city_weather["sys"]["country"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        description = city_weather['weather'][0]['description']
        
        city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        
        city_data.append({"City": city.title(),
                          "Country": city_country,
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Current Description":description })

    except:
        print("City not found. Skipping...")
        pass

print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
# Tally the number of cities in the city_data array of dictionaries
len(city_data)
```

```python
# Convert the array of dictionaries to a Pandas DataFrame.
city_data_df = pd.DataFrame(city_data)
city_data_df.head(10)
```

```python
# Create CSV file of DF
csv_weather_file = "weatherpy_database.csv"
# Export CSV file 
city_data_df.to_csv(csv_weather_file, index_label="City_ID")
```
